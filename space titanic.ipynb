```python
orig_train_X = train
orig_train_X = orig_train_X.drop("PassengerId", axis=1)
orig_train_X = orig_train_X.drop("Name", axis=1)
orig_train_X = orig_train_X.drop("Transported", axis=1)
orig_train_X
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="most_frequent")
asdf = imputer.fit_transform(orig_train_X)
orig_train_X=pd.DataFrame(asdf,index=orig_train_X.index,columns=orig_train_X.columns)
```

```python
import re

# 정규 표현식으로 Cabin을 각각 쪼개어준다.
pattern = r'([A-Z0-9]+)/([A-Z0-9]+)/([A-Z0-9]+)'
regex = re.compile(pattern,flags=re.IGNORECASE)

orig_train_X["c1"]=None
orig_train_X["c2"]=None
orig_train_X["c3"]=None

for idx in range(0,len(orig_train_X.index)):
    string = orig_train_X.loc[idx,"Cabin"]
    m = regex.match(string)
    qwer = m.groups()
    orig_train_X.loc[idx,"c1"]= qwer[0]
    orig_train_X.loc[idx,"c2"]= int(qwer[1])
    orig_train_X.loc[idx,"c3"]= qwer[2]
orig_train_X = orig_train_X.drop("Cabin", axis=1)
orig_train_X
```

```python
from sklearn.model_selection import StratifiedShuffleSplit as SSS

# n_splits는 몇 번 섞을지 결정
# StratifiedShuffleSplit이 알아서 각 계층에서 동일한 비율을 가지고 샘플을 뽑게 해준다.
split = SSS(n_splits=1, test_size=0.275, random_state=42)
for train_index, test_index in split.split(orig_train_X,orig_train_X["CryoSleep"]):
    train_X = orig_train_X.loc[train_index]
    train_y = orig_train_y.loc[train_index]
    val_X = orig_train_X.loc[test_index]
    val_y = orig_train_y.loc[test_index]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder


num_attribs = ["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck","c2"]
one_hot_attribs = ["HomePlanet", "Destination", "CryoSleep", "VIP","c3"]
ordi_attribs = ["c1"]

ordi_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="most_frequent")),
    ('ordi',OrdinalEncoder()),
    ('std_scaler', StandardScaler())
])

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])

one_hot_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="most_frequent")),
    ('one-hot', OneHotEncoder())
])

full_pipeline = ColumnTransformer([
    ("num",num_pipeline, num_attribs),
    ("cat", one_hot_pipeline,one_hot_attribs),
    ("ordi",ordi_pipeline,ordi_attribs)
])

train_X = full_pipeline.fit_transform(train_X)
train_X=train_X.astype('float')
val_X = full_pipeline.fit_transform(val_X)
val_X=val_X.astype('float')
```
